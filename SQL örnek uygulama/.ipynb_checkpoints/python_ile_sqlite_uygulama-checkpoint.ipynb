13. USA'de yaşayan müşterilerin kazandırdığı toplam kazancı bulunuz.

```python
pd.read_sql("select * from Orders limit 3;",conn)
```

```python
pd.read_sql("select * from Customers limit 3;",conn)
```

```python
pd.read_sql("select*from Products limit 3;",conn)
```

```python
pd.read_sql("select*from OrderDetails limit 3;",conn)
```

```python
pd.read_sql("""
select sum(p.Price*Quantity) as toplam_tutar
from Orders o
left join OrderDetails od on o.OrderID=od.OrderID
left join Products p on od.ProductID=p.ProductID
left join Customers c on o.CustomerID=c.CustomerID where c.Country='USA';""",conn)
```

14. Products tablosundaki CategoryID'lerin isimlerini bulunuz ve sonrasında kategori başına ortalama ürün fiyatını gösteriniz.

```python
pd.read_sql("""select c.CategoryName, avg(p.Price) 
from Products p 
inner join Categories c on p.CategoryID=c.CategoryID 
group by CategoryName order by Price desc;""",conn)
```

15. Çalışanları (EmployeeID) isimleri ile birlikte yaptıkları toplam satışlara göre sıralayınız.

```python
pd.read_sql("select name from sqlite_master where type='table';",conn)
```

```python
pd.read_sql("select * from Employees limit 10;",conn)
```

```python
pd.read_sql("select * from Orders limit 10;",conn)
```

```python
pd.read_sql("""
select e.FirstName as ad, e.LastName as soyad, count(o.OrderID) as toplam_satis 
from Orders o inner join Employees e on o.EmployeeID=e.EmployeeID 
group by ad,soyad order by toplam_satis desc;""",conn)
```

17. Almanya ya da USA'den verilen siparişlerin kategorilerine göre ortalama fiyatları nelerdir?

```python
pd.read_sql("""select ca.CategoryName, cu.Country, avg(p.Price) 
from Orders o 
left join OrderDetails od on o.OrderID=od.OrderID
left join Products p on od.ProductID=p.ProductID
left join Categories ca on p.CategoryID=ca.CategoryID 
left join Customers cu on o.CustomerID=cu.CustomerID
where Country in ('Germany','USA') group by CategoryName;
""",conn)
```

18. Haziran Temmuz Ağustos aylarında verilen siparişlerin ortalama fiyatı nedir?

```python
pd.read_sql("""
select avg(p.Price) from Products p
left join OrderDetails od on p.ProductID=od.ProductID
left join Orders o on od.OrderID=o.OrderID
where OrderDate like '%06%' or OrderDate like '%07%' or OrderDate like '%08%'""",conn)
```

19. Müşterilerin 1997 yılına ait siparişlerinin maksimum miktarlarını bulup müşterilerin isimleri ile birlikte büyükten küçüğe sıralayınız.

```python
pd.read_sql("""
select cu.CustomerName, max(o.OrderID) as maks 
from Products p 
left join OrderDetails od on p.ProductID=od.ProductID
left join Orders o on od.OrderId=o.OrderID
left join Customers cu on o.CustomerID=cu.CustomerID
where OrderDate like '%1997'
group by CustomerName order by maks desc""",conn)
```

20. Sipariş yılı 1997 olan siparişleri alan çalışanları sipariş alma sayılarına göre sıralayınız.

```python
pd.read_sql("""
SELECT COUNT(*) n, em.FirstName 
FROM Orders 
LEFT JOIN Employees em on Orders.EmployeeID = em.EmployeeID 
WHERE OrderDate like '%1997%' 
GROUP BY em.FirstName 
ORDER by n DESC;""",conn)
```
